# 无向图

```python
import networkx as nx
G=nx.Graph()
G.add_nodes_from(['v1', 'v2', 'v3', 'v4', 'v5'])
G.add_edges_from([('v1', 'v2'), ('v1', 'v3'), ('v2', 'v4'), ('v2', 'v5'), ('v3', 'v5'), ('v4', 'v5')])
print(nx.is_directed(G))
print(nx.is_weighted(G))
nx.draw(G, with_labels=True, pos=nx.spring_layout(G))
```

```python
A=nx.adjacency_matrix(G)
print(A)
print(A.todense())
```

```python
# 度矩阵
import numpy as np
D=np.diag(list(dict(nx.degree(G)).values()))
print(D)
```

```python
list(G.neighbors('v1'))
```

```python
print(G.degree())
print(G.nodes())
print(G.edges())
```

# 有向图

```python
G2=nx.DiGraph()
G2.add_nodes_from(['v1', 'v2', 'v3', 'v4', 'v5'])
G2.add_edges_from([('v1', 'v2'), ('v1', 'v3'), ('v2', 'v4'), ('v2', 'v5'), ('v3', 'v5'), ('v4', 'v5')])
print(nx.is_directed(G2))
print(nx.is_weighted(G2))
nx.draw(G2, with_labels=True, pos=nx.spring_layout(G))
```

```python
print(G2.in_degree())
print(G2.out_degree())
```
